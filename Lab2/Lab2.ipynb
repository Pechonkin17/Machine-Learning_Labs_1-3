Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.switch_backend('TkAgg')
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.tree import plot_tree
```

Download all datasets

```python
diabet_dataset_1 = pd.read_csv('diabet_datasets/diabetes_012_health_indicators_BRFSS2015.csv')
print(diabet_dataset_1)
```

```python
diabet_dataset_1.info()
```

```python
diabet_dataset_2 = pd.read_csv('diabet_datasets/diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
print(diabet_dataset_2)
```

```python
diabet_dataset_3 = pd.read_csv('diabet_datasets/diabetes_binary_health_indicators_BRFSS2015.csv')
print(diabet_dataset_3)
```

First dataset has thе values: 0, 1, 2. We need to replace 2 to 1, cos 2 another datasets are binary and 1 and 2 are types of diabet

```python
diabet_dataset_1.replace(2, 1, inplace=True)
print(diabet_dataset_1)
```

Rename column cos 2 datasets are binary, but third is not binary: 0 - no diabet, 1 - first type of diabet, 2 - second type of diabet

```python
diabet_dataset_1.rename(columns={'Diabetes_012': 'Diabetes_binary'}, inplace=True)
print(diabet_dataset_1)
```

Connection of three datasets

```python
merged_dataset = pd.concat([diabet_dataset_1, diabet_dataset_2, diabet_dataset_3], ignore_index=True)
print(merged_dataset.info())
```

Write columns with digital values

```python
# бажано б було замінити всі категоріяльні колонки на цифрові. Є 2 методи label, onehot - мб краще створює 2 додаткові колонки true, false в залежності скільки варіантів даної ознаки.
columns = ['Diabetes_binary', 'BMI', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
                   'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Income']
```

Replacing NaN values ​​with the arithmetic mean by column

```python
merged_dataset[columns] = merged_dataset[columns].fillna(merged_dataset[columns].mean())

categorical_columns = ['Diabetes_binary', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 'HeartDiseaseorAttack',
                       'PhysActivity', 'Sex', 'Education']

merged_dataset[categorical_columns] = merged_dataset[categorical_columns].astype('category')
```

Divide the training, validation and test dataset

```python
train_data, temp_data = train_test_split(merged_dataset, test_size=0.3, random_state=17)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=17)
```

Output of training, validation and test data sets

```python
print("Train data shape:", train_data.shape)
print("Validation data shape:", val_data.shape)
print("Test data shape:", test_data.shape)
```

```python
merged_dataset['Diabetes_binary'] = merged_dataset['Diabetes_binary'].astype('category')

# Побудова стовпчастої діаграми для відображення кількості людей з діабетом
plt.figure(figsize=(8, 6))
merged_dataset['Diabetes_binary'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')
plt.title('Number of People with Diabetes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Побудова кругової діаграми для відображення кількості людей з діабетом
plt.figure(figsize=(6, 6))
merged_dataset['Diabetes_binary'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Percentage of People with Diabetes')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
numeric_columns = merged_dataset.select_dtypes(include=np.number).columns.tolist()

# Функція для обрахунку медіани, середнього значення і моди
def calculate_statistics(column_name):
    mean_value = merged_dataset[column_name].mean()
    median_value = merged_dataset[column_name].median()
    mode_value = merged_dataset[column_name].mode().values[0] if not merged_dataset[column_name].mode().empty else np.nan
    return mean_value, median_value, mode_value

# Створення DataFrame для збереження результатів обрахунків
statistics_df = pd.DataFrame(columns=['mean', 'median', 'mode'])

# Обрахунок статистики для кожного числового стовпця
for column in numeric_columns:
    mean_val, median_val, mode_val = calculate_statistics(column)
    statistics_df.loc[column] = [mean_val, median_val, mode_val]

# Виведення результатів
print("Summary Statistics:")
print(statistics_df)
```

Нормалізація значень

Looking for the average of the columns

```python
numeric_columns = ['BMI', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
                   'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Income']
```

looking for the standard deviation by column

```python
means = train_data[numeric_columns].mean()
stds = train_data[numeric_columns].std()
```

```python
print("Means:\n", means)
print("\n""Standard deviations:\n", stds)
```

Normalize the data according to the formula

```python
normalized_train_data = train_data.copy()
normalized_train_data[numeric_columns] = (train_data[numeric_columns] - means) / stds

normalized_val_data = val_data.copy()
normalized_val_data[numeric_columns] = (val_data[numeric_columns] - means) / stds

normalized_test_data = test_data.copy()
normalized_test_data[numeric_columns] = (test_data[numeric_columns] - means) / stds
```

```python
features = ['BMI', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
                   'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Income']

# Х - дані які потрібні для передбачення, а Y - цільова колонка яку передбачаємо
X_train = normalized_train_data[features]
noisy_train_data = normalized_train_data.copy()
y_train = normalized_train_data['Diabetes_binary']

X_val = normalized_val_data[features]
y_val = normalized_val_data['Diabetes_binary']

X_test = normalized_test_data[features]
y_test = normalized_test_data['Diabetes_binary']

normalized_train_data.dropna(subset=['Diabetes_binary'], inplace=True)
X_train = normalized_train_data[features]
y_train = normalized_train_data['Diabetes_binary']
```

Building three different decision trees with different depths

```python
depths = [4, 6, 8]
models = []
for depth in depths:
    model = DecisionTreeClassifier(max_depth=depth, random_state=17)
    model.fit(X_train, y_train)
    models.append(model)

best_model = None
best_mse = float('inf')
```

Choose the best model for validation data

```python
for i, model in enumerate(models):
    y_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)
    print(f"Decision Tree with depth {depths[i]}: MSE = {mse}")
    if mse < best_mse:
        best_mse = mse
        best_model = model

y_pred_test = best_model.predict(X_test)
mse_test = mean_squared_error(y_test, y_pred_test)
print(f"Best Model: Decision Tree with depth {best_model.max_depth}, MSE = {best_mse}")
```

Visualization of the training model with the best depth

```python
plt.figure(figsize=(10, 6))
plot_tree(best_model, feature_names=features, filled=True)
plt.title(f"Best Decision Tree (depth {best_model.max_depth})")
plt.show()
```

Error output

```python
test_mse_best_model = mean_squared_error(y_test, best_model.predict(X_test))
print(f"Test MSE for the best model: {test_mse_best_model}")

test_mse = mean_squared_error(y_test, best_model.predict(X_test))
val_mse = mean_squared_error(y_val, best_model.predict(X_val))

print(f"Test MSE: {test_mse}")
print(f"Validation MSE: {val_mse}")

test_mse_best_model = mean_squared_error(y_test, best_model.predict(X_test))
print(f"Test MSE for the best model: {test_mse_best_model}")
```

PCA - shows which components have the greatest influence on the target trait (the greater the variance, the greater the influence)

Training PCA

```python
features = ['BMI', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
                   'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Income']

pca = PCA()
pca.fit(noisy_train_data[features])
```

Restore the data, that is, denoise

```python
# Reconstruct the data
reconstructed_data = pca.inverse_transform(pca.transform(noisy_train_data[features]))

# Perform PCA for denoising
pca_denoising = PCA(n_components=10)  # Choose the number of components for denoising
pca_denoising.fit(noisy_train_data[features])

# Denoise the data
denoised_data = pca_denoising.inverse_transform(pca_denoising.transform(noisy_train_data[features]))

depths = [4, 6, 8]
models_reconstructed = []
models_denoised = []

for depth in depths:
    # Train models on reconstructed data
    model_reconstructed = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model_reconstructed.fit(reconstructed_data, y_train)
    models_reconstructed.append(model_reconstructed)

    # Train models on denoised data
    model_denoised = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model_denoised.fit(denoised_data, y_train)
    models_denoised.append(model_denoised)

# Evaluate the models
for i, model in enumerate(models_reconstructed):
    y_pred_reconstructed = model.predict(X_test)
    mse_reconstructed = mean_squared_error(y_test, y_pred_reconstructed)
    print(f"Decision Tree with depth {depths[i]} on Reconstructed Data: MSE = {mse_reconstructed}")

for i, model in enumerate(models_denoised):
    y_pred_denoised = model.predict(X_test)
    mse_denoised = mean_squared_error(y_test, y_pred_denoised)
    print(f"Decision Tree with depth {depths[i]} on Denoised Data: MSE = {mse_denoised}")
```

```python
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
import numpy as np

# Perform PCA for data reconstruction
pca = PCA()
pca.fit(noisy_train_data[features])

# Calculate cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Determine the number of components for 95% variance explained
n_components_95 = np.argmax(cumulative_variance_ratio >= 0.95) + 1

# Perform PCA for denoising with n_components_95
pca_denoising = PCA(n_components=n_components_95)
pca_denoising.fit(noisy_train_data[features])

# Denoise the data
denoised_data = pca_denoising.inverse_transform(pca_denoising.transform(noisy_train_data[features]))

depths = [4, 6, 8]
models_denoised = []

for depth in depths:
    # Train models on denoised data
    model_denoised = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model_denoised.fit(denoised_data, y_train)
    models_denoised.append(model_denoised)

# Evaluate the models
for i, model in enumerate(models_denoised):
    y_pred_denoised = model.predict(X_test)
    mse_denoised = mean_squared_error(y_test, y_pred_denoised)
    print(f"Decision Tree with depth {depths[i]} on Denoised Data: MSE = {mse_denoised}")
```
